```python
# 1. Сколько у нас пользователей, которые совершили покупку только один раз? 
```

```python
# посмотрела, что в olist_customers находится уникальный идентификатор пользователя, 
# а в olist_orders только позаказный идентификатор пользователя
# при этом в обеих таблицах одинаковое число строчек, поэтому можно сделать иннер джоин по customer_id, ничего не потеряв

# за покупку стала считать заказ, у которого есть время подтверждения оплаты заказа (order_approved_at) и статус delivered,
# потому что до доставки пользователь может отменить покупку
```

```python
purchases = olist_customers.merge(olist_orders, on = 'customer_id', how = 'inner')
purchases.head()
```

```python
# сначала посмотрела, что есть заказы, у которых не указано время оплаты

purchases.order_approved_at.isna().unique()
```

```python
# посмотрела какие вообще могут быть статусы заказов

purchases.order_status.unique()
```

```python
# сделала минидатафрейм для подсчета, куда отобрала только тех пользователей, у которых указано время подтверждения заказа 
# и он доставлен
# посмотрела сколько таких строк

delivered_purchases = purchases[['customer_id', 'customer_unique_id','order_status','order_approved_at']]\
                      .query('order_status == "delivered"').dropna()
delivered_purchases.shape
```

```python
# сначала сгруппировала по customer_unique_id, чтобы посчитать сколько покупок совершил каждый уникальный пользователь,
# потом сгруппировала по этому количеству покупок и посмотрела сколько пользователей совершило такое количество покупок

unique_user_purchases = delivered_purchases\
                                            .groupby('customer_unique_id', as_index = False)\
                                            .agg({'customer_id' : 'count'})\
                                            .rename(columns = {'customer_id' : 'purchases_amount'})\
                                            .groupby('purchases_amount',as_index = False)\
                                            .agg({'customer_unique_id' : 'count'})\
                                            .rename(columns = {'customer_unique_id' : 'unique_users_amount'})
unique_user_purchases               
```

```python
# проверка

(unique_user_purchases.purchases_amount * unique_user_purchases.unique_users_amount).sum()
```

```python
# 96464 пользователей совершили покупку всего 1 раз, если считать покупку успешной при оплате и завершенной доставке товара 
```

```python
# 2. Сколько заказов в месяц в среднем не доставляется по разным причинам (вывести детализацию по причинам)? (5 баллов)
```

```python
# order_delivered_customer_date это дата, когда заказ был доставлен покупателю
# order_delivered_carrier_date - дата, когда заказ доставлен логистической службе
# полагаю, если эти даты не указаны, значит заказ не доставляется
# еще не брала те, где статус заказа явно указан как delivered

undelivered_orders = olist_orders.loc[(olist_orders.order_delivered_customer_date.isnull())
                                    & (olist_orders.order_status != 'delivered') 
                                    & (olist_orders.order_delivered_carrier_date.isnull())]
undelivered_orders
```

```python
# посмотрела типы данных, я заранее поменяла тип order_purchase_timestamp на datetime64
# взяла время создания заказа order_purchase_timestamp для работы с датой

undelivered_orders.dtypes
```

```python
# вытащила из даты только месяц

month = undelivered_orders.order_purchase_timestamp.dt.month_name()
month
```

```python
# вытащила из даты только год

year = undelivered_orders.order_purchase_timestamp.dt.year
year
```

```python
# лепила между собой с помощью индексов

# на этот вариант пандас начал ругаться что меняю срез датафрейма
# undelivered_orders['order_month'] = undelivered_orders.order_purchase_timestamp.dt.month_name() 


undelivered_orders = undelivered_orders.merge(month, left_index=True, right_index=True)\
                                   .rename(columns = {'order_purchase_timestamp_y' : 'order_month'})
undelivered_orders
```

```python
# не получилось слепить за раз, потому что там возникли трудности с повторяющимися названиями колонок 
# и я не стала себя путать

# undelivered_orders['order_year'] = undelivered_orders.order_purchase_timestamp.dt.year 

undelivered_orders = undelivered_orders.merge(year, left_index=True, right_index=True)\
                                   .rename(columns = {'order_purchase_timestamp' : 'order_year'})
undelivered_orders
```

```python
# посмотрела число заказов, которые не доставляются в срезе по их статусу, году и месяцу

undelivered_orders_by_time = undelivered_orders\
                                            .groupby(['order_status', 'order_year', 'order_month',], as_index = False)\
                                            .agg({'order_id':'count'})\
                                            .rename(columns={'order_id' : 'undelivered_orders'})
undelivered_orders_by_time.head()
```

```python
# сгруппировала по статусу для детализации и посчитала среднее число заказов, которые не доставляются, по месяцам
# за 2016-2017-2018

# т.е. беру среднее число по месяцам за три года, чтоб посмотреть тенденцию

undelivered_orders_by_time_mean = undelivered_orders_by_time\
                                    .groupby(['order_status','order_month'], as_index = False)\
                                    .agg({'undelivered_orders' : 'mean'})\
                                    .rename(columns = {'undelivered_orders' : 'undelivered_orders_mean'})
undelivered_orders_by_time_mean
```

```python
# визуализировала

ax = sns.lineplot(data = undelivered_orders_by_time_mean, x = 'order_month', y = 'undelivered_orders_mean', hue = 'order_status')
ax.set_title('Undelivered orders mean by month\n')
ax.set_xlabel('\nMonth')
ax.set_ylabel('Undelivered orders mean\n')
ax.set_xticklabels(undelivered_orders_by_time_mean.order_month, rotation = 45)
sns.despine()
```

```python
# 3. По каждому товару определить, в какой день недели товар чаще всего покупается.
```

```python
# здесь уже за покупку стала считать заказ, у которого есть время подтверждения оплаты заказа (order_approved_at),
# потому что пользователь что-то заказал и даже оплатил это в какой-то день недели, 
# сейчас не учитываю отменил ли он его потом

orders_by_day = olist_orders.loc[olist_orders.order_approved_at.notnull()]
orders_by_day
```

```python
olist_order_items.head()
```

```python
# посмотрела, что есть пользователи, у которых несколько товаров в заказе

olist_order_items.groupby('order_id',as_index= False).agg({'product_id':'count'}).head(20)
```

```python
# поэтому, когда мерджила в результате декартова умножения, стало больше строк

orders_by_day = orders_by_day.merge(olist_order_items, on = 'order_id', how = 'inner')
orders_by_day.head()
```

```python
# посмотрела как это выглядит

orders_by_day.loc[orders_by_day.order_id == '0008288aa423d2a3f00fcb17cd7d8719']
```

```python
orders_by_day.dtypes
```

```python
# вытащила день недели и так же прилепила по индексу

weekday = orders_by_day.order_purchase_timestamp.dt.day_name()

orders_by_day = orders_by_day.merge(weekday, left_index=True, right_index=True)\
                                   .rename(columns = {'order_purchase_timestamp_y' : 'order_weekday'})
orders_by_day
```

```python
# посчитала количество всех заказов каждого товара по дням недели
# один товар могут заказывать в разные дни

all_orders_by_day = orders_by_day\
                                    .groupby(['product_id','order_weekday'], as_index = False)\
                                    .agg({'order_id' : 'count'})\
                                    .rename(columns = {'order_id' : 'order_count'})
all_orders_by_day
```

```python
# нашла количество каждого товара, который покупали во все дни недели

all_orders = all_orders_by_day\
                                .groupby('product_id', as_index = False)\
                                .order_count.sum()\
                                .rename(columns = {'order_count' : 'orders_sum'})
all_orders
```

```python
# примерджила общее количество покупок товара за все дни ко всем покупкам товаров по дням
# и посчитала долю покупок в конкретный день от всех покупок товара за все дни

all_orders_by_day = all_orders_by_day.merge(all_orders, on = 'product_id')
all_orders_by_day = all_orders_by_day.assign(pecentage = all_orders_by_day.order_count/all_orders_by_day.orders_sum)
```

```python
# оставила только те товары и день недели, где доля покупок этого товара была максимальной

max_orders_by_day = all_orders_by_day\
                                     .groupby('product_id', as_index = False).max()

max_orders_by_day
```

```python
# проверила сколько уникальных товаров было в начальном датафрейме, когда смерджила

orders_by_day.product_id.nunique()
```

```python
# а в датафрейме с максимальным числом покупок остался только вторник, потому что в этот день было максимальное количество

max_orders_by_day.query('product_id == "fffdb2d0ec8d6a61f0a0a0db3f25b441"')
```

```python
# Сколько у каждого из пользователей в среднем покупок в неделю (по месяцам)? 
# Не стоит забывать, что внутри месяца может быть не целое количество недель. 
# Например, в ноябре 2021 года 4,28 недели. И внутри метрики это нужно учесть. 
```

```python
# добавила колонку где посчитала колиечство недель в месяце

purchases['week_in_month'] = (purchases.order_purchase_timestamp.dt.days_in_month / 7).round(2)
```

```python
# сделала датафрейм по покупкам в ответе на первый вопрос
# снова отобрала только те заказы, которые доставлены и есть время подтверждения оплаты

ready_purchases = purchases.loc[(purchases.order_status == 'delivered') & (purchases.order_approved_at.notnull())]
ready_purchases.head()
```

```python
ready_purchases.dtypes
```

```python
# опять вытащила все месяцы и прилепила их по индексам
# не могу придумать ничего умнее С:

month = ready_purchases.order_purchase_timestamp.dt.month_name()
ready_purchases = ready_purchases.merge(month, left_index=True, right_index=True)\
                                   .rename(columns = {'order_purchase_timestamp_y' : 'order_month'})
ready_purchases.head()
```

```python
# сгруппировала по пользователю, месяцу, когда он совершил покупку, и количеству недель в этом месяце
# (на группировку количество недель не влияет, потому что тесно связано с месяцем)
# посчитала какое количество покупок пользователь совершил в конкретный месяц
# поделила число покупок в конкретный месяц на количество недель в этом месяце

purchases_mean = ready_purchases\
                                .groupby(['customer_unique_id', 'order_month', 'week_in_month'],as_index = False)\
                                .order_id.count()\
                                .rename(columns = {'order_id' : 'purchases_count'})

purchases_mean['purchases_mean'] = (purchases_mean.purchases_count / purchases_mean.week_in_month).round(2)
purchases_mean
```

```python
# ради интереса посмотрела какое вообще бывает количество покупок в месяц

purchases_mean.purchases_count.unique()
```

```python
# Используя pandas, проведи когортный анализ пользователей. 
# В период с января по декабрь выяви когорту с самым высоким retention на 3й месяц.
```

```python
# отобрала оплаченные доставленные заказы

paid_purchases =  purchases.loc[(purchases.order_approved_at.notnull()) & (purchases.order_status == 'delivered')]
```

```python
paid_purchases.dtypes
```

```python
paid_purchases.shape
```

```python
# проверила что повторов нет

paid_purchases.drop_duplicates()
```

```python
print('min date of payment purchase is', paid_purchases.order_purchase_timestamp.min())
print('max date of payment purchase is', paid_purchases.order_purchase_timestamp.max())
```

```python
# вытащила месяц и год, когда пользователь заплатил за свой заказ

payment_month = paid_purchases['order_purchase_timestamp'].dt.to_period('M')

paid_purchases = paid_purchases.merge(payment_month, left_index=True, right_index=True)\
                                   .rename(columns = {'order_purchase_timestamp_y' : 'payment_month'})
paid_purchases
```

```python
# сгруппировала по пользователям и взяла для каждого дату, когда он совершил самую первую покупку (оплату)
# отнесла пользователя к когорте в зависимости от года и месяца его первой покупки


cohort = paid_purchases.groupby('customer_unique_id').order_purchase_timestamp_x \
                         .transform('min') \
                         .dt.to_period('M')

paid_purchases = paid_purchases.merge(cohort, left_index=True, right_index=True)\
                                   .rename(columns = {'order_purchase_timestamp_x_y' : 'cohort'})
paid_purchases
```

```python
# сгруппировала по когорте пользователей и месяцу оплаты (например, пользователь из 1 когорты мог совершить несколько покупок)
# посчитала сколько покупок по месяцам совершают пользователи каждой когорты

cohort_df = paid_purchases.groupby(['cohort', 'payment_month'], as_index=False) \
                  .agg({'customer_unique_id': pd.Series.nunique})\
                  .rename(columns = {'customer_unique_id' : 'user_count'})  


# вытаскиваю сколько месяцев прошло между первой покупкой пользовтаеля и его конкретной покупкой
# После f = attrgetter('name.first', 'name.last') вызов f(b) возвращает (b.name.first, b.name.last)
# f = attrgetter('n')
# f(cohort_df.period_number[0]) = 0


cohort_df['period_number'] = (cohort_df.payment_month - cohort_df.cohort) \
                             .apply(attrgetter('n'))
```

```python
cohort_df
```

```python
# делаю сводную таблицу, где указано количество пользователей из каждой когорты, которые совершили еще покупки спустя какое-то
# количество месяцев после своей первой покупки

cohort_pivot = cohort_df.pivot_table(index='cohort',
                                     columns='period_number',
                                     values='user_count')
cohort_pivot
```

```python
# беру первую колонку (где месяц 0) и использую в качестве 100%, далее каждую ячейку каждой строчки делю на соответствующую
# цифру из колонки 0, чтобы посчитать долю пользователей этой когорты, совершающих повторную покупку, 
# в конкретный месяц от первоначального числа пользователей в этой когорте

cohort_size = cohort_pivot.iloc[:, 0]
retention_matrix = cohort_pivot.divide(cohort_size, axis=0)
retention_matrix
```

```python

ur_style = (retention_matrix
            .style
            .set_caption('User retention by cohort')  # добавляем подпись
            .background_gradient(cmap='magma')  # раскрашиваем ячейки по столбцам
            .highlight_null('white')  # делаем белый фон для значений NaN
            .format("{:.2%}", na_rep=""))  # числа форматируем как проценты, NaN заменяем на пустоту
ur_style
```

```python
# Часто для качественного анализа аудитории использую подходы, основанные на сегментации. 
# Используя python, построй RFM-сегментацию пользователей, чтобы качественно оценить свою аудиторию. 
# В кластеризации можешь выбрать следующие метрики: 
# R - время от последней покупки пользователя до текущей даты,
# F - суммарное количество покупок у пользователя за всё время, 
# M - сумма покупок за всё время. Подробно опиши, как ты создавал кластеры. 
# Для каждого RFM-сегмента построй границы метрик recency, frequency и monetary для интерпретации этих кластеров. 
# Пример такого описания: RFM-сегмент 132 (recency=1, frequency=3, monetary=2) имеет границы метрик recency от 130 до 500 дней, 
# frequency от 2 до 5 заказов в неделю, monetary от 1780 до 3560 рублей в неделю.
```
